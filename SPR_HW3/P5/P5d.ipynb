```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
mean1 = np.array([-3, 0])
sigma1 = np.array([[1.5, 1], [1, 1.5]])
mean2 = np.array([3, 0])
sigma2 = np.array([[1.5, -1], [-1, 1.5]])
```

```python
xlist = np.linspace(-10.0, 10.0, 100)
ylist = np.linspace(-10.0, 10.0, 100)
X, Y = np.meshgrid(xlist, ylist)
pos = np.array([Y,X]).T
rv1 = st.multivariate_normal(mean1, sigma1)
sample1 = rv1.rvs(500)
Z1 = rv1.pdf(pos)
rv2 = st.multivariate_normal(mean2, sigma2)
sample2 = rv2.rvs(500)
Z2 = rv2.pdf(pos)
plt.figure(figsize = (8,8))
plt.contour(X, Y, Z1)
plt.contour(X, Y, Z2)
plt.title('')
plt.ylabel("x2")
plt.xlabel("x1")
plt.scatter(sample1[:,0], sample1[:,1])
plt.scatter(sample2[:,0], sample2[:,1])
plt.title("Part d")
plt.show()
```

```python
w = np.array([-2, 0])
w0 = np.array([[5/9], [0]])*np.log(3/10)
Z = np.dot(w, np.append(np.expand_dims(X.ravel(),axis=1),np.expand_dims(Y.ravel(), axis=1), axis=1).T-w0)

plt.figure(figsize = (8,8))
plt.contour(X, Y, Z1)
plt.contour(X, Y, Z2)
plt.contour(X, Y, Z.reshape(X.shape), levels = 0)
plt.title('')
plt.ylabel("x2")
plt.xlabel("x1")
plt.scatter(sample1[:,0], sample1[:,1])
plt.scatter(sample2[:,0], sample2[:,1])
plt.title("Part d")
plt.show()
```
