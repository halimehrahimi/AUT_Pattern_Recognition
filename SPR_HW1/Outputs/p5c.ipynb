```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

# c1.

```python
#Initialization
sigma1 = np.array([[2, 0], [0, 2]])
mean1 = np.array([-5, 5])
#Using Meshgrid for Contour
xlist = np.linspace(-10.0, 10.0, 100)
ylist = np.linspace(-10.0, 10.0, 100)
X, Y = np.meshgrid(xlist, ylist)
#Calculating pdf for all the points in the meshgrid
pos = np.array([Y,X]).T
rv1 = st.multivariate_normal(mean1, sigma1)
sample1 = rv1.rvs(500)
Z = rv1.pdf(pos)
#Plotting the samples and contour
plt.figure(figsize = (10,10))
plt.contour(X, Y, Z)
plt.title('Part c - c1\nCircle in the Upper Left')
plt.ylabel("x2")
plt.xlabel("x1")
plt.scatter(sample1[:,0], sample1[:,1])
plt.show()
```
