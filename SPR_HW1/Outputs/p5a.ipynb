```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
#Given Initialization
mean = 5
sigma = np.array([1, 2, 3])
x = np.linspace(0,500,500)
#Random Samples
sample1 = st.multivariate_normal.rvs(mean, sigma[0], 500)
sample2 = st.multivariate_normal.rvs(mean, sigma[1], 500)
sample3 = st.multivariate_normal.rvs(mean, sigma[2], 500)
```

```python
#Plotting the Scatter Plots and Histograms
fig, ax = plt.subplots(3,2,sharex='col', sharey='row' ,figsize=(15,15))
ax[0,0].set_title('Part a\nScatter Plots\nSigma=1')
ax[0,0].scatter(x, sample1)
ax[0,1].set_title('Part a\nHistograms\nSigma=1')
ax[0,1].hist(sample1, 100)
ax[1,0].set_title('Sigma=2')
ax[1,0].scatter(x, sample2)
ax[1,1].set_title('Sigma=2')
ax[1,1].hist(sample2, 100)
ax[2,0].set_title('Sigma=3')
ax[2,0].scatter(x, sample3)
ax[2,1].set_title('Sigma=2')
ax[2,1].hist(sample3, 100)
plt.show()
```
