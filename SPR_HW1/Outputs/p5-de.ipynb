```python
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
```

# d.

```python
ro = [-0.99, -0.5, 0.5, 0.99]
```

```python
#Printing the Equations
for p in ro:
    print("d^2 = %s ((2*y-4)^2 + (x-3)^2 - 4(%s)(y-2)(x-3))" % (1/(4*(1-p**2)),p))
```

# e

```python
#Using Meshgrid for Contour
xrange = np.linspace(-10, 10, 200)
yrange = np.linspace(-10, 10, 200)
y, x = np.meshgrid(xrange,yrange)

d2list = [4,9,16]

for p in ro:
    i=0
    fig, axs = plt.subplots(3,1, figsize=(8,24))
    #the distribution
    pos = np.array([x,y]).T
    sigma = np.array([[1, p*2], [p*2, 4]])
    mean = np.array([2, 3])
    rv1 = st.multivariate_normal(mean, sigma)
    Z = rv1.pdf(pos)
    #the d2 contours
    for d2 in d2list:
        z = ((1/(4*(1-p**2)))*((2*y-4)**2+(x-3)**2 -(4*p*(y-2)*(x-3))))-d2
        axs[i].set_title("Contour for p = "+str(p)+" , d2 = "+str(d2))
        axs[i].contour(y,x,z,levels=0,colors='white')
        axs[i].contourf(y,x,Z)
        axs[i].set_xlabel('x1')
        axs[i].set_ylabel('x2')
        i+=1
    plt.subplots_adjust(hspace=0.4)
plt.show()
```
